# Manejo de datos nulos

```python
#Creamos un df con valores nulos
dict = {'Col1': [1,2,3,np.nan],
        'Col2': [4,np.nan,6,7],
        'Col3': ['a','b','c',None]}

df = pd.DataFrame(dict)
df
```

```python
#Funcion que regresa True si es nulo o False en caso contrario
df.isnull()
```

```python
#podemos transofrmarlo a 0 y 1
df.isnull()*1
```

```python
#Reemplazar los valores nulos por un valor o una funcion
df.fillna('Missing')
```

```python
df.fillna(df.mean())
```

```python
#llenar los valores nulos por columna suponiendo que ésta sigue una serie
df.interpolate()
```

```python
#borrar las filas con valores nulos
df.dropna()
```
