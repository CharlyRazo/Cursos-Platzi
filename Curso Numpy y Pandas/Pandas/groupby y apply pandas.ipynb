# Groupby
Agrupamientos por columna en la que vamos a trabajar sobre registros que tengan en comun dicha columna y poder aplicarle una funcion de agregacion: count(),sum(), mean(), max(), min(), etc.

```python
df_books = pd.read_csv('bestsellers-with-categories.csv')
df_books.head()
```

```python
#la columna 'Author' deja de ser una columna y se convierte en indice
df_books.groupby('Author').count()
```

```python
#Author al ser indice podemos aplicar loc
df_books.groupby('Author').count().loc['William Davis']
```

```python
#para que autor sea una columna debemos colocar reset_index()
df_books.groupby('Author').count().reset_index()
```

```python
#podemos aplicar mas de una funcion de agregacion a la vez
df_books.groupby('Year').agg(['min','max'])
```

```python
#podemos especificar diversas funciones de agregacion para cada columna
df_books.groupby('Author').agg({'Reviews':['min','max'], 'User Rating': 'sum'})
```

```python
#podemos agrupa por dos columnas
df_books.groupby(['Author', 'Year']).count()
```

# Apply
Nos permite crear funciones y aplicarlas sobre el dataframe entero o sobre una columna y es mucho mas rapido que aplicar un ciclo for sobre todas las filas o columnas

```python
df_books.head()
```

```python
#definimos una funcion
def two_times(value):
    return value * 2
```

```python
df_books['Rating 2'] = df_books['User Rating'].apply(two_times)
df_books
```

```python
#lo mas habitual es usar funciones de tipo lambda
df_books['Rating_3'] = df_books['User Rating'].apply(lambda x: x*3)
df_books
```

```python
df_books = pd.read_csv('bestsellers-with-categories.csv')
#aplicamos una funcion a todo el dataframe
df_books['Rating_2'] = df_books.apply(lambda x: x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'], axis=1)
df_books
```
