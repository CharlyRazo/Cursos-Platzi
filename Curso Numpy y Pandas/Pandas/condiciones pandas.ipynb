# Condiciones
Al igual que numpy, regresa una serie o df con booleanos

```python
#A una serie o df que satisfaga una condicion se le llama tambien: mascara
mayor_a_2016 = df_books['Year'] >2016
mayor_a_2016
```

```python
#ocupamos la mascara para solo visualizar aquellos renglones que satisfagan la condicion
df_books[mayor_a_2016]
```

```python
#aquellos de aÃ±o mayor a 2016 y de genero de ciencia ficcion
genre_fiction = df_books['Genre'] == 'Fiction'
df_books[genre_fiction & mayor_a_2016]
```

```python
#para negar condicion
df_books[~mayor_a_2016]
```
