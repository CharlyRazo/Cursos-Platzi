# Subplot

Permite la creación de subgráficas en una misma imagen y escoger que gráfica irá en que posición.

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# parametros (# de filas, # de columnas, indice)
plt.subplot(1,2,1) #grafico con indice 1
plt.plot(x,y,'r--')

plt.subplot(1,2,2) #grafico con indice 2
plt.hist(y)
plt.show()
```

```python
plt.subplot(2,1,1) #grafico con indice 1
plt.plot(x,y,'b--')
plt.plot(y,x,'r:')

plt.subplot(2,1,2) #grafico con indice 2
plt.pie(y)
plt.show()
```

# Object Oriented
Es una forma de crear gráficos de matplotlib. A diferencia de los pyplot básicos, éste requiere mayor código, pero con las ventajas:
- Mayor personalización
- Más amigable con imágenes de múltiples gráficos

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# metodo orientado a objetos, para ello creamos un objeto 'fig'
# este objeto sera nuestro lienzo para realizar las graficas
fig = plt.figure()
```

```python
# a partir del objeto definimos las gráficas con el nombre 'axes'
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.3,0.6]) #los ultimos dos numeros permiten customizar el tamaño
axes.plot(x,y,'b')
fig.show() #este comando sirve para mostrar la figura sin el objeto
```

```python
# añadimos una segunda figura
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.6,0.4,0.3]) # los primeros dos digitos indican la posicion de la figura
axes.plot(x,y,'b')
axes2.plot(x,y,'r')
axes2.set_facecolor('gray')
fig.show()
```

Vemos que podemos customizar las gráficas de forma independiente, por el hecho de ser dos objetos independientes, siendo ésta la ventaja principal del método orientado a objetos.

# Subplots con Object Oriented

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
#creamos un lienzo y nombramos una figura llamada axes en una sola linea
fig, axes = plt.subplots(nrows=1,ncols=2)
```

```python
#vamos nombrando cada graficas con axes[i] como si fuera un objeto iterable
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
plt.show()
```

```python
#podemos definir las graficas directamente con un arreglo o tupla
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
plt.show()
```

```python
# para cuando tengamos varias filas y columnas podemos acceder a 
# cada grafica de la siguiente forma
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,np.cos(x), 'b')
axes[0,1].plot(x,np.sin(x), 'g')
axes[0,2].plot(x,np.tan(x), 'y')
axes[0,3].plot(x,np.cos(x)**2, 'r')
```

```python
# otra forma de hacerlo
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4)
ax1.plot(x,np.cos(x), 'b')
ax2.plot(x,np.sin(x), 'g')
ax3.plot(x,np.tan(x), 'y')
ax4.plot(x,np.cos(x)**2, 'r')
fig.tight_layout() #coloca espacios entre graficos para que sea más estético
```

# Leyendas, etiquetas, titulos, tamaño

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2, figsize = (10,5)) #tamaño
axes[0].plot(x,y, label = '$sin(x)$') 
axes[0].set_title('Relacion X-Y') #titulo del subplot
axes[0].set_xlabel('X') #titulo en los ejes
axes[0].set_xlabel('Y')
axes[0].legend() #permite que se muestre el label o etiqueta

axes[1].plot(y,x, label = '$arcsin(x)$')
axes[1].set_title('Relacion Y-X') #titulo del subplot
axes[1].set_xlabel('Y')
axes[1].set_xlabel('X')
axes[1].legend()

fig.suptitle('Este es un título global', fontsize = 20)  #titulo global
plt.show()
```

# Barplot
Gráfica importante para visualizar variables categóricas.

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY'] #var categorica
population = [1000, 800, 900, 1000, 300] #var numerica
```

```python
plt.style.use('default') #volvemos a las graficas por default
plt.bar(country,population, width = 0.2) #ancho de barras
plt.show()
```

```python
plt.bar(country,population, color = ['red', 'blue', 'green']) #volor de barras
plt.xticks(np.arange(5),('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45) #leyendas
plt.show()
```

```python
#grafica de barras horizontal
plt.barh(country,population)
plt.show()
```

# Otras graficas

```python
data = np.random.randint(1,50,100)
```

Histograma: Ideal para mostrar la distribución de un dataset y saber como analizarlo

```python
#bins = Numero de barras (rangos o intervalos en que divide el dataset)
#histtype = tipo de histograma
plt.hist(data, bins=15, width= 2)
plt.show()
```

```python
#bins = Numero de barras (rangos o intervalos en que divide el dataset)
#histtype = formato de histograma
plt.hist(data, bins=15,histtype = 'step')
plt.show()
```

Boxplot: Otra forma de conocer como se conoce la distribución, suele ser mas ideal para identificar quantiles y los datos outliers

```python
#vert = si la caja es horizontal o vertical
#patch_artist = colorea el boxplot para mejorar la visualizacion
#notch,= da una forma poligonal para visualizar mejor la mediana
plt.boxplot(data, vert = False, patch_artist = True, notch=True)
plt.show()
```

```python
#vamos a agregar outliers
outliers = np.array([150,170,200])
data = np.append(data, outliers)

plt.boxplot(data, vert = False, patch_artist = True, notch=True)
plt.show()
```

Scatterplot: Ideal para visualizar la dependencia entre dos o mas variables

```python
N = 50
x = np.random.rand(N) #N numeros entre 0 y 1
y = np.random.rand(N)
area = (30*x)**2
colors = np.random.rand(N)
```

```python
#s = representa el tamaño de los amrcadores y permite hacer grafico de burbujas

plt.scatter(x,y, s=area, c = colors, marker = 'p', alpha = 0.5)
plt.show()
```
