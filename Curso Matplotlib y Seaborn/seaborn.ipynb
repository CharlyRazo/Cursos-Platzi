# Distribuciones
Seaborn ofrece algunas gráficas u opciones que te permiten trabajar con distribuciones orientadas a tipos de datos numéricos. Veamos algunas de ellas.

```python
#improtamos un dataset de sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set(style = None, palette = 'bright' )

#para visualizar una distribucion, la grafica que debemos realizar es el histograma
# cumulative = es para visualizar la func de distribucion cumulativa
# stat = es para que en el eje y diga la probabilidad, porcentaje en vez de la frecuencia 
# multiple = puedes hacer el grafico de barras apiladas (stack), las barras unas al lado de otras (dodge)
#            o por default una detras de otra (layer)
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative= False, hue = 'sex',
            stat = 'count', multiple = 'dodge')
plt.show()
```

```python
#distribucion de densidad continua
sns.kdeplot(data = tips,x='tip',hue = 'sex', cumulative = False, bw_adjust=1)
plt.show()
```

```python
#la funcion displot permite realizar las graficas anteriores especificando 
# con kind si queremos un histograma (hist) o una funcion de densidad (kde)
sns.displot(data = tips,x='tip',hue = 'sex', kind = 'kde', multiple = 'stack')
plt.show()
```

# Visualizacion de datos categóricos

Seaborn ofrece diferentes gráficos para graficar datos categóricos o variables de texto. A continuación veremos algunos de ellos:

```python
tips.head()
```

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
# El siguietne grafico es una combinacion entre scatterplot y grafico de barras  
# donde en el eje X seleccionas la categoria y en el eje Y seleccionas la metrica 
# numérica en la que deseas comparar. A este grafico se le llama grafico de enejambre

plt.figure(figsize=(6,8))
sns.stripplot(data=tips, x='day', y = 'total_bill', hue='sex', dodge = True)
plt.show()
```

```python
# similar al grafico anterior solo que el eje horizontal tambien indica la acumulacion 
# de datos en ese punto
plt.figure(figsize=(6,8))
sns.swarmplot(data=tips, x='day', y = 'total_bill', hue='sex', dodge = True)
sns.boxplot(data = tips, x = 'day', y = 'total_bill')
plt.show()
```

```python
# los boxplot permiten visualizar la distribucion estadistica por percentiles
# en cada categoria de la variable
sns.boxplot(data=tips, x='day', y = 'total_bill', hue='sex', dodge = True)
plt.show()
```

```python
#podemos combinar dos graficas en una
sns.boxplot(data=tips, x='day', y = 'total_bill', hue='sex', dodge = True)
sns.swarmplot(data=tips, x='day', y = 'total_bill', hue='sex', dodge = True, color = 'black', marker = '<')
plt.show()
```

```python
# grafica de violin: similar al boxplot pero la concentracion de los datos lo representa
# a traves del ancho de la figura
# con la variable split, combina las dos separaciones dadas por Hue  en una sola
sns.violinplot(data=tips, x='day', y = 'total_bill', hue='sex', dodge = True, split=True)
plt.show()
```

```python
#la grafica de violin tambien se puede con variable numerica
sns.violinplot(data=tips, y = 'total_bill')
plt.show()
```

```python
# catplot define de forma automatica el grafico
# con kind podemos especificar el grafico y con
# col podemos tener mas graficos de acuerdo a otra categoria
sns.catplot(data=tips, x='day', y='total_bill', hue = 'sex', dodge = True, kind = 'box', col = 'time')
plt.show()
```

# Graficos de tipo relacion (scatterplot)
Seaborn maneja los gráficos de relación entre distintas variables numéricas para hacer visualizaciones. Veamos los distintos gráficos que tiene Seaborn para trabajar este tipo de relaciones.

```python
tips.head()
```

```python
#style =  permite otra separacion dada una variable categorica por medio de marcadores
#         similar a lo que hace "hue" con colores 
sns.scatterplot(data=tips, x='total_bill', y= 'tip', hue = 'day', style='time', size = 'size')
plt.legend(loc = 'center', bbox_to_anchor = (1.12,0.5))
plt.show()
```

```python
# podemos generar un diccionario donde definamos marcadares dada una variable categorica
markers = {'Lunch':'D', 'Dinner': 's'}
sns.scatterplot(data=tips, x='total_bill', y= 'tip', hue = 'day', style='time', 
                size = 'size', markers = markers)
plt.legend(loc = 'center', bbox_to_anchor = (1.12,0.5))
plt.show()
```

```python
# grafico de linea
sns.lineplot(data=tips, x='total_bill', y= 'tip', hue = 'day', style='time', 
            size = 'size')
plt.legend(loc = 'center', bbox_to_anchor = (1.12,0.5))
plt.show()
```

```python
# La siguiente figura es similar al Catplot para datos categoricos pero para datos numericos
# con kind podemos especificar el tipo de grafico y con col obtener mas graficos dada una 
# variable categorica
sns.relplot(data=tips, x='total_bill', y= 'tip', hue = 'day', style='time', 
            size = 'size', col ='time')
plt.show()
```

# Joinplot y Pairplot
Nos permite combinar mas de un tipo de gráfico para realizar comparativas de forma más fácil.

```python
# Joinplot cobina dos graficas, las de dispersion e histograma
# con el fin de conocer la distribucion de dos variables
# y su corelacion entre ellos
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
plt.show()
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', kind = 'hist', 
            marginal_ticks = True, marginal_kws = dict(bins=25,fill=False, multiple = 'dodge'))
plt.show()
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', kind = 'kde')
plt.show()
```

```python
# Pairplot toma todas las variables numericas y las cruza entre si para compararlas
# con scatterplots e histrogramas
sns.pairplot(data = tips)
plt.show()
```

```python
# algunas customizaciones
# corner: elimina la informacion duplicada quedandose con solo un triangulo de la matriz
sns.pairplot(data = tips, hue = 'sex', corner = True , kind = 'scatter')
plt.show()
```

# Heatmap
Este es otro tipo de grafico muy utilizado en ciencia de datos, ya que correlaciona todas las variables numéricas de un dataset de forma muy visual.

```python
tips.head()
```

```python
#correlacion de pearson entre las variables numericas
tips.corr()
```

```python
sns.heatmap(tips.corr())
plt.show()
```

```python
# annot = permite colocar los valores de la correlacion en cada cuadro
# cmap : mapa de colores
# linewidths = ancho en lineas de separacion
# vmin  y vmax = permiten visualizar con una barra el calor de cada correlacion de mejor manera
sns.heatmap(tips.corr(), annot = True, cmap = 'coolwarm', linewidths = 5, vmin = 0.5, vmax=1, cbar = True)
plt.show()
```
